# Exploratory Data Analysis of the Pima Indians Diabetes Dataset

## Introduction

The dataset contains medical diagnostic measurements for 768 female patients of Pima Indian heritage.  
The goal is to predict whether a patient has diabetes (binary outcome).

The features present in the dataset are:

- **Pregnancies**: Number of times pregnant  
- **Glucose**: Plasma glucose concentration (mg/dL)  
- **BloodPressure**: Diastolic blood pressure (mm Hg)  
- **SkinThickness**: Triceps skinfold thickness (mm)  
- **Insulin**: 2-Hour serum insulin (µU/mL)  
- **BMI**: Body mass index (weight in kg / (height in m)^2)  
- **DiabetesPedigreeFunction**: A function which scores likelihood of diabetes based on family history  
- **Age**: Age in years  

The target variable is:

- **Outcome**: Indicates whether the patient has diabetes (0 = No, 1 = Yes).  

```python
# importing pandas, and other necessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Define column names (from UCI Pima Indians Diabetes dataset)
col_names = [
    "Pregnancies", 
    "Glucose", 
    "BloodPressure", 
    "SkinThickness", 
    "Insulin", 
    "BMI", 
    "DiabetesPedigreeFunction", 
    "Age", 
    "Outcome"
]

# Load dataset
df = pd.read_csv("pima-indians-diabetes.csv", header=None, names=col_names)
```

## 2. Target Variable Distribution (Xác định tính chất)
Check class balance of the outcome variable (0 = no diabetes, 1 = diabetes).  

```python
# Check class balance (assuming column 'Outcome' or 'Diabetes') 
label_col = None
for cand in ['Outcome','Diabetes','target','Class','HasDiabetes','diabetes']:
    if cand in df.columns:
        label_col = cand
        break

if label_col:
    print(f"Label column detected: {label_col}\n")
    
    counts = df[label_col].value_counts()
    proportions = df[label_col].value_counts(normalize=True) * 100
    
    print("Class distribution:")
    for cls in counts.index:
        print(f"  Class {cls}: {counts[cls]} samples ({proportions[cls]:.2f}%)")
else:
    print("No label column detected.")
```

## 3. Suspicious Zeros and Missing Values (Xác định giá trị thiếu, và mẫu bất thường)
Some features (Glucose, BloodPressure, SkinThickness, Insulin, BMI) cannot realistically be 0.  
We will check and replace them with NaN.  

```python
# Missing / suspicious zeros 
def suspicious_zero_summary(df):
    candidates = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI','Age']
    present = [c for c in candidates if c in df.columns]

    stats = []
    for c in present:
        zeros = (df[c] == 0).sum()
        stats.append([c, zeros, round(zeros/len(df)*100, 2)])
    return pd.DataFrame(stats, columns=["Feature", "Zeros", "Percentage (%)"])

zero_stats = suspicious_zero_summary(df)
print("\nSuspicious Zeros Summary:\n")
print(zero_stats.to_string(index=False))

# Replace zeros with NaN
to_nan = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
df[to_nan] = df[to_nan].replace(0, np.nan)

# Show missing values after replacement
missing_stats = df[to_nan].isna().sum().reset_index()
missing_stats.columns = ["Feature", "Missing Values"]
print("\nMissing Values After Replacement:\n")
print(missing_stats.to_string(index=False))
```

## 4. Univariate Analysis (Phân tích đơn biến)
- Histograms to observe feature distributions  
- Descriptive stats including missing values  
- Boxplots to detect outliers  

```python
# Univariate analysis
for col in df.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(6,3))
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"Histogram: {col}")
    plt.tight_layout()
    plt.show()

# Descriptive stats with missing counts
num = df.select_dtypes(include=[np.number]).columns
desc = df[num].describe().T
desc['missing'] = df[num].isna().sum()
desc['missing(%)'] = (df[num].isna().sum() / len(df) * 100).round(2)

print("\nDescriptive Statistics with Missing Values:\n")
print(desc.round(2).to_string(index=True))
```

## 6. Outlier Detection (Xác định ngoại lệ)
Use the Interquartile Range (IQR) method to detect outliers. 

```python
# Detect outliers by IQR
def detect_outliers_iqr(series, k=1.5):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k*iqr
    upper = q3 + k*iqr
    return series[(series < lower) | (series > upper)]

outlier_summary = {}
for c in num:
    out = detect_outliers_iqr(df[c].dropna())
    outlier_summary[c] = {
        "outliers": len(out),
        "outliers(%)": round(len(out) / len(df) * 100, 2)
    }

out_df = pd.DataFrame(outlier_summary).T

print("\nOutliers Detection (IQR method):\n")
print(out_df.to_string(index=True))
```

## 7. Missing Value Treatment (Xử lý giá trị khuyết)
We can impute missing values using the median.

```python
# Missing value treatment suggestions (numeric)
print("Missing percent:")
print(df.isna().mean().round(3))

# Simple imputation example: median
df_imputed = df.copy()
for c in to_nan:
    df_imputed[c] = df_imputed[c].fillna(df_imputed[c].median())
```

## 8. Class-wise Summary (Tóm tắt theo nhóm)
Compare median feature values between diabetic and non-diabetic groups.  

```python
# Check class-wise summaries
if label_col:
    print(df_imputed.groupby(label_col)[num].median().T)
```
